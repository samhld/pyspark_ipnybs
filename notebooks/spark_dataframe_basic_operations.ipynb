```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('ops').getOrCreate()
```

```
df = spark.read.csv('/FileStore/tables/appl_stock.csv',inferSchema=True,header=True)
```

```
# using SQL, just cause
df.filter("close < 500").show()
```

```
df.filter(df[4] < 500).select('Open','Close').show(10)
```

```
df.filter((df['Close'] < 200) & (df['Open'] > 200)).show()
```

```
# Same as above but instead it's "'Close' less than 200 AND NOT 'Open' is greater than 200"...shows 5 because output was larger
df.filter((df['Close'] < 200) & ~ (df['Open'] > 200)).show(5)
```

```
# Check which days had a 'Low' of a certain value
df.filter(df['Low'] == 197.16).select('Date','Low').show()
```
