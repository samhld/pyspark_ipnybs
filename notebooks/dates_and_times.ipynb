```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('dates').getOrCreate()
```

```
df = spark.read.csv('/FileStore/tables/appl_stock.csv',header=True,inferSchema=True)
```

```
df.show()
```

```
from pyspark.sql.functions import (dayofmonth,
                                   hour,
                                   dayofyear,
                                   month,
                                   year,
                                   weekofyear,
                                   format_number,
                                   date_format)
```

```
df.select(dayofmonth(df['Date'])).show(2)
```

```
df.select(month(df['Date'])).show(2)
```

```
sliceable = df.select(weekofyear(df['Date']))
```

```
sliceable[0][0]
```

```
df.select('Open')
```

```
df.withColumn('Year',year(df['Date'])).show(4)
```

```
newdf = df.withColumn('Year',year(df['Date']))
```

```
# Mean by year
newdf.groupBy("Year").mean().show(4)
```

```
result = newdf.groupBy("Year").mean().select(["Year","avg(Close)"])
```

```
result.show(4)
```

```
new_result = result.withColumnRenamed("avg(Close)","Avg close per year")
```

```
new_result.select(["Year",format_number("Avg close per year",2).alias("Avg close per year")]).show()
```
